## Assignment 1: Spiral Order of a Matrix

Given a `m × n` matrix, return the elements in spiral order.

```python
def print_spiral_order(matrix):
    if not matrix or not matrix[0]:
        return []

    if hasattr(matrix, "tolist"):
        matrix = matrix.tolist()

    m, n = len(matrix), len(matrix[0])
    result = []

    top, bottom = 0, m - 1
    left, right = 0, n - 1

    while top <= bottom and left <= right:
        for c in range(left, right + 1):
            result.append(matrix[top][c])
        top += 1

        for r in range(top, bottom + 1):
            result.append(matrix[r][right])
        right -= 1

        if top <= bottom:
            for c in range(right, left - 1, -1):
                result.append(matrix[bottom][c])
            bottom -= 1

        if left <= right:
            for r in range(bottom, top - 1, -1):
                result.append(matrix[r][left])
            left += 1

    return result
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print(print_spiral_order(matrix))
```

## Assignment 2: Word Frequency Counter

```python
def count_word_frequency(sentence):
    freq = {}
    words = sentence.lower().split()

    for word in words:
        clean = ''.join(c for c in word if c.isalnum())
        if clean:
            freq[clean] = freq.get(clean, 0) + 1

    return freq
```

```python
print(count_word_frequency("This is a test. This test is simple."))
```

## Assignment 3: Palindrome Check

```python
def is_palindrome_iterative(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
```

```python
def is_palindrome_recursive(s):
    if len(s) <= 1:
        return True
    if s[0] != s[-1]:
        return False
    return is_palindrome_recursive(s[1:-1])
```

```python
print(is_palindrome_iterative("radar"))
print(is_palindrome_recursive("world"))
```

## Assignment 4: Prime Number Check

```python
def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6

    return True
```

```python
print(is_prime(2))
print(is_prime(97))
print(is_prime(20))
```
