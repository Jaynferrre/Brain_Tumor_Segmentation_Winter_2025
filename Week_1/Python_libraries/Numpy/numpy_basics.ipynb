## Assignment 1

1. Create a random NumPy array of shape (3, 6)
2. Add `[0, 0, 2, 4, 5, 3]` to each row
3. Reshape to (9, 2)
4. Take transpose
5. Compute mean
6. Find indices of elements greater than mean

```python
import numpy as np

# 1. Create an array of 18 random elements of shape (3, 6)
arr = np.random.randint(0, 20, size=(3, 6))
print("Original array (3,6):\n", arr)

# 2. Add the array [0, 0, 2, 4, 5, 3] to each row
row_to_add = np.array([0, 0, 2, 4, 5, 3])
modified_arr = arr + row_to_add
print("\nArray after adding to each row:\n", modified_arr)

# 3. Reshape it to a (9, 2) array
reshaped_arr = modified_arr.reshape(9, 2)
print("\nReshaped array (9,2):\n", reshaped_arr)

# 4. Take its transpose
transposed_arr = reshaped_arr.T
print("\nTransposed array:\n", transposed_arr)

# 5. Find the mean of the transposed array
mean_of_arr = np.mean(transposed_arr)
print(f"\nMean of the transposed array: {mean_of_arr:.2f}")

# 6. Find locations where elements are greater than the mean
locations = np.where(transposed_arr > mean_of_arr)

# 7. Print the locations
print("\nLocations (indices) where elements are greater than the mean:")
for r, c in zip(locations[0], locations[1]):
    print(f"  Element at ({r}, {c}): {transposed_arr[r, c]}")
```

## Assignment 2

1. Generate a Poisson-distributed array (λ = 5)
2. Compute mean and standard deviation
3. Center the array
4. Normalize using standard deviation

```python
import numpy as np

# 1. Create an array of length 20 following a Poisson distribution
poisson_array = np.random.poisson(lam=5, size=20)
print("Original Poisson Array:\n", poisson_array)

# 2. Calculate mean and standard deviation
mean_poisson = np.mean(poisson_array)
std_poisson = np.std(poisson_array)
print(f"\nMean of Poisson array: {mean_poisson:.2f}")
print(f"Standard Deviation of Poisson array: {std_poisson:.2f}")

# 3. Center the array around its mean
centered_array = poisson_array - mean_poisson
print("\nCentered Array (mean subtracted):\n", centered_array)

# 4. Normalize the array
normalized_array = centered_array / std_poisson if std_poisson != 0 else centered_array
print("\nNormalized Array (divided by std dev):\n", normalized_array)
```
